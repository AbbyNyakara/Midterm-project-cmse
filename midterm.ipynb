## Exploratory Data Analysis

```python
# Total dosses administered for different age-groups
administered_doses = df.groupby('Age Group')['Residents Vaccinated'].sum().sort_values().reset_index()
administered_doses
```

```python
sns.barplot(x='Residents Vaccinated', y="Age Group", data=administered_doses)
```

#### Give the barplot axis labels, a title, same color and ability to hover and see values

```python
## Doses administered by the Race/Ethnicity
administered_by_race = df.groupby('Race/Ethnicity')['Residents Vaccinated'].sum().sort_values().reset_index()
administered_by_race
```

```python
## Donut Chart
labels = administered_by_race['Race/Ethnicity']
sizes = administered_by_race['Residents Vaccinated']

fig, ax = plt.subplots(figsize=(6, 16), subplot_kw=dict(aspect="equal"))
ax.pie(sizes, labels=labels, autopct='%1.1f%%')
ax.set_title("Vaccinations by Race")

plt.show()
```
